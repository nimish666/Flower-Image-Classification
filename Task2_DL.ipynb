```python
import cv2
import os
from skimage import feature

def read_images_and_labels(img_dir, label_file):
    # Read the labels from the label file
    with open(label_file) as f:
        labels = [word.lower().strip() for word in f.read().split("\n") if word.strip()]

    # Create a mapping of label to its index

    labels = [eval(i) for i in labels]
    list_set = set(labels)
    corrected_labels = (list(list_set))


    final_data=[]

    for label in set(corrected_labels):
        matching_files = [f for f in os.listdir(img_dir) if f.startswith(f"{label}_image_") and f.endswith('.jpg')]

        for matching_file in matching_files:
            image_path = os.path.join(img_dir, matching_file)
            image = cv2.imread(image_path, cv2.IMREAD_COLOR)

            if image is not None:
                final_data.append((image,label-1))

    return final_data
```

```python
import torch
import numpy as np
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms, models
import torch.optim as optim
from torch import nn
import matplotlib.pyplot as plt
from PIL import Image

train_images=[]
train_labels=[]
for image,label in images_data:
    train_images.append(image)
    train_labels.append(label)

train_images=np.array(train_images)
train_labels=np.array(train_labels)

val_images=[]
val_labels=[]
for img,lbl in val_data:
    val_images.append(img)
    val_labels.append(lbl)

val_images=np.array(val_images)
val_labels=np.array(val_labels)

class FlowerDataset(Dataset):
    def __init__(self, images, labels, transform=None):
        self.images = images
        self.labels = labels
        self.transform = transform

    def __len__(self):
        return len(self.images)

    def __getitem__(self, idx):
        image = self.images[idx]
        label = self.labels[idx]

        image = Image.fromarray((image * 255).astype(np.uint8))  

        if self.transform:
            image = self.transform(image)  

        return image, label

transform = transforms.Compose([
    transforms.Resize((224, 224)),  
    transforms.ToTensor(),  
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])  
])


# Create DataLoader for training data
train_dataset = FlowerDataset(train_images, train_labels, transform=transform)
train_loader = DataLoader(train_dataset, batch_size=16, shuffle=True)

# Create DataLoader for validation data
val_dataset = FlowerDataset(val_images, val_labels, transform=transform)
val_loader = DataLoader(val_dataset, batch_size=16, shuffle=False)

num_classes = len(np.unique(train_labels))

# Load Pretrained DenseNet Model
densenet = models.densenet121(pretrained=True)  

# Modify the final fully connected layer to match your number of classes
densenet.classifier = nn.Linear(densenet.classifier.in_features, num_classes)

device = torch.device('cpu')  
densenet.to(device)

# Define Optimizer and Loss Function
optimizer = optim.Adam(densenet.parameters(), lr=1e-4)
criterion = nn.CrossEntropyLoss()

# Training function
def train_epoch(model, data_loader, optimizer, criterion, device):
    model.train()
    running_loss = 0
    correct = 0
    total = 0

    for images, labels in data_loader:
        images, labels = images.to(device), labels.to(device)

        # Cast labels to Long (required for CrossEntropyLoss)
        labels = labels.long()

        # Forward pass
        outputs = model(images)
        loss = criterion(outputs, labels)

        # Backward pass and optimization
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        # Track accuracy
        _, preds = torch.max(outputs, 1)
        correct += (preds == labels).sum().item()
        total += labels.size(0)

        running_loss += loss.item()

    epoch_loss = running_loss / len(data_loader)
    epoch_acc = correct / total
    return epoch_loss, epoch_acc

# Validation function
def evaluate(model, data_loader, criterion, device):
    model.eval()  # Set model to evaluation mode
    running_loss = 0
    correct = 0
    total = 0

    with torch.no_grad():  
        for images, labels in data_loader:
            images, labels = images.to(device), labels.to(device)

            # Cast labels to Long (required for CrossEntropyLoss)
            labels = labels.long()

            # Forward pass
            outputs = model(images)
            loss = criterion(outputs, labels)

            # Track accuracy
            _, preds = torch.max(outputs, 1)  # Get the predicted class index
            correct += (preds == labels).sum().item()
            total += labels.size(0)

            running_loss += loss.item()

    # Calculate and return average loss and accuracy for the epoch
    epoch_loss = running_loss / len(data_loader)
    epoch_acc = correct / total
    return epoch_loss, epoch_acc


# Early stopping class
class EarlyStopping:
    def __init__(self, patience=2, delta=0):
        self.patience = patience  
        self.delta = delta  
        self.best_loss = None  
        self.counter = 0  
        self.early_stop = False  

    def __call__(self, val_loss, model):
        if self.best_loss is None:
            self.best_loss = val_loss  
        elif val_loss < self.best_loss - self.delta:
            self.best_loss = val_loss  
            self.counter = 0  
        else:
            self.counter += 1  
            if self.counter >= self.patience:
                self.early_stop = True  


early_stopping = EarlyStopping(patience=2, delta=0.01)


epochs = 10
for epoch in range(epochs):
    print(f"Epoch {epoch+1}/{epochs}")

    # Train the model
    train_loss, train_acc = train_epoch(densenet, train_loader, optimizer, criterion, device)
    print(f"Train Loss: {train_loss:.4f}, Train Accuracy: {train_acc:.4f}")

    # Evaluate the model on the validation set
    val_loss, val_acc = evaluate(densenet, val_loader, criterion, device)
    print(f"Validation Loss: {val_loss:.4f}, Validation Accuracy: {val_acc:.4f}")

    # Check for early stopping
    early_stopping(val_loss, densenet)

    if early_stopping.early_stop:
        print(f"Early stopping at epoch {epoch+1}")
        break

# Step 5: Visualize some predictions
def visualize_predictions(images, labels, predictions, class_names):
    images = images.cpu().numpy().transpose(0, 2, 3, 1)
    fig, axes = plt.subplots(1, 5, figsize=(15, 3))
    for ax, img, lbl, pred in zip(axes, images, labels, predictions):
        ax.imshow(img)
        ax.set_title(f"True: {class_names[lbl]}, Pred: {class_names[pred]}")
        ax.axis('off')
    plt.show()

# Get a batch of validation images and predictions
images, labels = next(iter(val_loader))
images, labels = images.to(device), labels.to(device)
outputs = densenet(images)
_, preds = torch.max(outputs, 1)

# Visualize the predictions
visualize_predictions(images, labels, preds, class_names=np.unique(labels).tolist())
```

```python
"""
 Use this cell to test the model.
 1.Set the path of folder with the test data in path_test
 2.Set the path of label file in path_test_label variable
 3.Then RUN this cell
 It will show accuracy and also the predictions
 Assumption: Data in test set is also named as label_image_{some_number} and labels for test set are given in a separate file
 Just like it was for validation set and training set.

"""
# Test Bench

import torch
import numpy as np
from torchvision import models, transforms
from PIL import Image
from torch.utils.data import DataLoader, TensorDataset
import os, cv2
from torch import nn
from sklearn.metrics import accuracy_score

# Define the transform to match the training process
transform = transforms.Compose([
    transforms.Resize((224, 224)),  
    transforms.ToTensor(),  
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])  
])

dev= 'cpu'
if torch.cuda.is_available():
    dev= 'cuda'
device = torch.device(dev)  


# Path of test set
path_test="/content/val_data"
# Path of test set labels
path_test_label="/content/val_labels.txt"


with open(path_test_label) as f:
        test_labels = [word.lower().strip() for word in f.read().split("\n") if word.strip()]

test_labels = [eval(i) for i in test_labels]
list_set = set(test_labels)
corrected_labels = (list(list_set))

X_test_pre=[]

for label in set(corrected_labels):

    matching_files = [f for f in os.listdir(path_test) if f.startswith(f"{label}_image_") and f.endswith('.jpg')]

    for matching_file in matching_files:
        image_path = os.path.join(path_test, matching_file)
        img = cv2.imread(image_path, cv2.IMREAD_COLOR)
        X_test_pre.append(img)

images_list = []
for image in X_test_pre:
    pil_image = Image.fromarray((image * 255).astype(np.uint8))
    transformed_image = transform(pil_image)
    images_list.append(transformed_image)


images_tensor = torch.stack(images_list)  
images_tensor = images_tensor.to(device)

# Define a DataLoader for batching
batch_size = 16  
dataset = TensorDataset(images_tensor)  
data_loader = DataLoader(dataset, batch_size=batch_size, shuffle=False)

# Function to predict the class indices
def predict_batch(data_loader, model, device):
    all_predictions = []

    # Set the model to evaluation mode
    model.eval()

    with torch.no_grad():  # No need to compute gradients for inference
        for images in data_loader:
            images = images[0]  # Get the images from the tuple (only one element)

            images = images.to(device)

            outputs = model(images)
            
            _, predicted_classes = torch.max(outputs, 1)

            # Convert to numpy and collect all predictions
            predictions = predicted_classes.cpu().numpy()
            all_predictions.extend(predictions)

    return all_predictions

# Load the saved model
densenet = models.densenet121(pretrained=False)
densenet.classifier = nn.Linear(densenet.classifier.in_features, 60)  # Ensure the classifier is correct
densenet.load_state_dict(torch.load('flower_model.pth', map_location=device))
densenet.to(device)

# Predict for all images in the batch
predicted_classes = predict_batch(data_loader, densenet, device)

adjusted_predictions = [pred + 1 for pred in predicted_classes]

# Print the first 10 predictions (class indices)
print(adjusted_predictions)

accuracy = accuracy_score(test_labels, adjusted_predictions)
print(f'Accuracy: {accuracy * 100:.2f}%')
```
